```python
import requests
from google.colab import userdata
city=input("Enter the city:").strip()
api_key=userdata.get('OPENWEATHER_KEY')
url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
response=requests.get(url)
data=response.json()
if data["cod"]!=200:
  print("Error",data.get("message","unknown error!"))
else:
  try:
     temperature=data["main"]["temp"]
     weather=data["weather"][0]["description"]
     humidity=data["main"]["humidity"]
     wind_speed=data["wind"]["speed"]
     print("\n"+"="*30)
     print("📍WEATHER REPORT:",city.upper())
     print("="*30)
     print("🌡Temperature:",temperature,"C")
     print("☁Condition:",weather)
     print("💧Humidity:",humidity,"%")
     print("💨wind Speed:",round(wind_speed*3.6,1),"km/h")
     print("="*30)
     if temperature>35:
      print("🔥Very hot weather.")
      print("Stay hydrated🥤")
     elif temperature<15:
      print("\n🥶Very cold weather")
      print("Wear layered clothing to trap body heat effectively🧥")
     else:
      print("\n😊Pleasant weather")
     print("="*30)
     if "rain"in weather:
      print("☔Take an umbrella")
     if "cloud" in weather.lower():
      print("☁Sky is cloudy")
     if wind_speed>5:
      print("🌬Windy weather")

  except Exception as e:
    print("Something went wrong",e)
```

```python
%%writefile app.py
import streamlit as st
import requests

st.title("🌦️ Weather Data Analyzer App")

api_key = "YOUR_API_KEY"

city = st.text_input("Enter a city name")

if city:

    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()

    if data["cod"] != 200:
        st.error(data.get("message", "Error occurred"))
    else:
        temp = data["main"]["temp"]
        weather = data["weather"][0]["description"]
        humidity = data["main"]["humidity"]
        wind = data["wind"]["speed"]

        st.subheader(f"📍 Weather in {city}")

        st.write("🌡️ Temperature:", temp, "°C")
        st.write("🌥️ Condition:", weather)
        st.write("💧 Humidity:", humidity, "%")
        st.write("🌬️ Wind Speed:", wind, "m/s")

        st.subheader("🧠 Weather Insight")

        if temp >= 35:
            st.write("🔥 Extremely Hot Weather")
        elif temp >= 25:
            st.write("🌤️ Warm Weather")
        elif temp >= 15:
            st.write("🧥 Cool Weather")
        else:
            st.write("❄️ Cold Weather")

        if "rain" in weather:
            st.write("🌧️ Take an umbrella")
        elif "clear" in weather:
            st.write("☀️ Good weather for going out")
```
