```python
import requests
from google.colab import userdata
city=input("Enter the city:").strip()
api_key=userdata.get('OPENWEATHER_KEY')
url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
response=requests.get(url)
data=response.json()
if data["cod"]!=200:
  print("Error",data.get("message","unknown error!"))
else:
  try:
     temperature=data["main"]["temp"]
     weather=data["weather"][0]["description"]
     humidity=data["main"]["humidity"]
     wind_speed=data["wind"]["speed"]
     print("\n"+"="*30)
     print("📍WEATHER REPORT:",city.upper())
     print("="*30)
     print("🌡Temperature:",temperature,"C")
     print("☁Condition:",weather)
     print("💧Humidity:",humidity,"%")
     print("💨wind Speed:",round(wind_speed*3.6,1),"km/h")
     print("="*30)
     if temperature>35:
      print("🔥Very hot weather.")
      print("Stay hydrated🥤")
     elif temperature<15:
      print("\n🥶Very cold weather")
      print("Wear layered clothing to trap body heat effectively🧥")
     else:
      print("\n😊Pleasant weather")
     print("="*30)
     if "rain"in weather:
      print("☔Take an umbrella")
     if "cloud" in weather.lower():
      print("☁Sky is cloudy")
     if wind_speed>5:
      print("🌬Windy weather")

  except Exception as e:
    print("Something went wrong",e)
```
